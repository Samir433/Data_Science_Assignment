```python
# combine all
df = transactions.merge(customers, on='CustomerID').merge(products, on='ProductID')
```

```python
df
```

```python
# calucalte the TotalSpend TotalSpend by each customer
customer_features = transactions.groupby('CustomerID').agg(
    TotalSpend=('TotalValue', 'sum'),
    TotalQuantity=('Quantity', 'sum')
).reset_index()
```

```python
customer_features
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
columns_to_scale = ['TotalSpend', 'TotalQuantity']
scaler = StandardScaler()
customer_features[columns_to_scale] = scaler.fit_transform(customer_features[columns_to_scale])
```

```python
customer_features
```

```python
# create picit table based on purchased procut by each cust
grouped_data = transactions.groupby(['CustomerID', 'ProductID'])['Quantity'].sum().reset_index()
pivoted_data = grouped_data.pivot(index='CustomerID', columns='ProductID', values='Quantity').fillna(0)
pivoted_data = pivoted_data.reset_index()
```

```python
pivoted_data
```

```python
pivoted_data
```

```python
# category_summary for the each cutomer 
category_dummies = pd.get_dummies(df['Category'], prefix='Category')
df_with_dummies = pd.concat([df[['CustomerID']], category_dummies], axis=1)
category_summary = df_with_dummies.groupby('CustomerID').sum().reset_index()
category_summary
```

```python
customer_features
```

```python
customers
```

```python
# merge all
merged_data = customers.merge(customer_features, on='CustomerID', how='left') \
                       .merge(pivoted_data.reset_index(), on='CustomerID', how='left') \
                       .merge(category_summary, on='CustomerID', how='left')
```

```python
merged_data = merged_data.drop(columns=['index'])
```

```python
merged_data = merged_data.fillna(0)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

numerical_data = merged_data.drop(columns=['CustomerID'])[:20]

cos_sim_matrix = cosine_similarity(numerical_data)

similar_customers = []

for i in range(len(cos_sim_matrix)):
    sim_scores = cos_sim_matrix[i]
    sim_list = [(merged_data['CustomerID'][j], sim_scores[j]) for j in range(len(sim_scores)) if i != j]
    sim_list_sorted = sorted(sim_list, key=lambda x: x[1], reverse=True)[:3]
    
    simi1 = sim_list_sorted[0] if len(sim_list_sorted) > 0 else (None, None)
    simi2 = sim_list_sorted[1] if len(sim_list_sorted) > 1 else (None, None)
    simi3 = sim_list_sorted[2] if len(sim_list_sorted) > 2 else (None, None)
    
    similar_customers.append((merged_data['CustomerID'][i], simi1[0], simi1[1], simi2[0], simi2[1], simi3[0], simi3[1]))

similarity_df = pd.DataFrame(similar_customers, columns=['cid', 'simi1_cid', 'simi1_score', 'simi2_cid', 'simi2_score', 'simi3_cid', 'simi3_score'])
```

```python
similarity_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
numerical_data = merged_data.drop(columns=['CustomerID'])[:20] 
cos_sim_matrix = cosine_similarity(numerical_data)

lookalikes = []

for i in range(20): 
    sim_scores = cos_sim_matrix[i]  
    sim_list = [(merged_data['CustomerID'][j], sim_scores[j]) for j in range(len(sim_scores)) if i != j]
    sim_list_sorted = sorted(sim_list, key=lambda x: x[1], reverse=True)[:3]
    
    lookalikes.append({
        'cust_id': merged_data['CustomerID'][i],
        'lookalikes': [(cust[0], cust[1]) for cust in sim_list_sorted] 
    })
lookalike_df = pd.DataFrame(lookalikes)
lookalike_df['lookalikes'] = lookalike_df['lookalikes'].apply(lambda x: str(x))
lookalike_df[['cust_id', 'lookalikes']].to_csv('Lookalike.csv', index=False)
```
